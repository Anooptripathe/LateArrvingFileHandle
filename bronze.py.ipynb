```python
from pyspark.sql.functions import current_timestamp, col

bronze_df = (
    spark.read
    .option("header",True)
    .format('csv')
    .load('/Volumes/assetbundles/default/xyz/Warehouse_and_Retail_Sales.csv')
    .select('*', '_metadata')
    
    .withColumn('ingestts', current_timestamp())
    .withColumn('filename', col('_metadata.file_name'))
)
```

```python
bronze_df = bronze_df.toDF(
    *[col.replace(" ", "_").replace("-", "_").replace("/", "_") for col in bronze_df.columns]
)
bronze_df.write \
    .option("mergeSchema", "true")\
    .format("delta") \
    .mode("append") \
    .partitionBy("event_date") \
    .saveAsTable("bronze_sales")
      
```
